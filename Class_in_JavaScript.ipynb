**Problem Statement 1: Create a Constructor for a Person**

**Write a JavaScript function constructor named Person that takes two parameters, name, and age, and Add a method to the prototype to display name and age. Then, create two instances of Person and display their names and ages.**

-



```
// Constructor function for Person
function Person(name, age) {
    this.name = name;
    this.age = age;
}

// Adding a method to the prototype to display name and age
Person.prototype.display = function() {
    console.log("Name: " + this.name + ", Age: " + this.age);
};

// Creating two instances of Person
let person1 = new Person("Alice", 25);
let person2 = new Person("Bob", 30);

// Displaying their names and ages
person1.display();  // Output: Name: Alice, Age: 25
person2.display();  // Output: Name: Bob, Age: 30
```

**Problem Statement 2: Implement a Bank Account**

**Create a constructor function named BankAccount that initializes a bank account with an initial balance. Include methods for depositing and withdrawing money from the account. Write code to demonstrate these operations on a bank account object.**

-  



```
// Constructor function for BankAccount
function BankAccount(initialBalance) {
    this.balance = initialBalance;
}

// Method to deposit money
BankAccount.prototype.deposit = function(amount) {
    if (amount > 0) {
        this.balance += amount;
        console.log("Deposited: $" + amount + ". New balance: $" + this.balance);
    } else {
        console.log("Deposit amount must be positive.");
    }
};

// Method to withdraw money
BankAccount.prototype.withdraw = function(amount) {
    if (amount > 0 && amount <= this.balance) {
        this.balance -= amount;
        console.log("Withdrew: $" + amount + ". New balance: $" + this.balance);
    } else if (amount > this.balance) {
        console.log("Insufficient funds. Current balance: $" + this.balance);
    } else {
        console.log("Withdrawal amount must be positive.");
    }
};

// Method to check balance (optional, for demonstration)
BankAccount.prototype.getBalance = function() {
    return this.balance;
};

// Demonstrating the operations
let account = new BankAccount(1000);  // Initial balance: $1000
console.log("Initial balance: $" + account.getBalance());

account.deposit(500);  // Deposit $500
account.withdraw(200); // Withdraw $200
account.withdraw(1500); // Attempt to withdraw more than balance
account.deposit(-100); // Attempt invalid deposit

console.log("Final balance: $" + account.getBalance());
```

**Problem Statement 5: Implement a Calculator Class**

**Create a class called Calculator that initializes two values value1 and value2 to store numbers. Add methods for add, subtract, multiply, and divide. Perform and show operations.**

-  


```
// Calculator class
class Calculator {
    constructor(value1, value2) {
        this.value1 = value1;
        this.value2 = value2;
    }

    // Method to add values
    add() {
        return this.value1 + this.value2;
    }

    // Method to subtract values
    subtract() {
        return this.value1 - this.value2;
    }

    // Method to multiply values
    multiply() {
        return this.value1 * this.value2;
    }

    // Method to divide values
    divide() {
        if (this.value2 !== 0) {
            return this.value1 / this.value2;
        } else {
            return "Cannot divide by zero";
        }
    }
}

// Create an instance of Calculator
const calc = new Calculator(20, 5);

// Perform operations
console.log("Addition:", calc.add());         // 25
console.log("Subtraction:", calc.subtract()); // 15
console.log("Multiplication:", calc.multiply()); // 100
console.log("Division:", calc.divide());      // 4
```

Problem Statement 6: Design a Class for a Geometric Shape

Design a class called Shape that can represent various geometric shapes. Implement subclasses for specific shapes like Circle, Rectangle, and RightTriangle. Each subclass should have properties of that shape (e.g..
Width, height for a rectangle) and methods for calculating the area and perimeter of the shape. Create instances of these shapes and calculate their areas and perimeters.


-  



```
// Base class for Shape
class Shape {
    constructor(type) {
        this.type = type;
    }
}

// Subclass for Circle
class Circle extends Shape {
    constructor(radius) {
        super("Circle");
        this.radius = radius;
    }

    area() {
        return Math.PI * this.radius * this.radius;
    }

    perimeter() {
        return 2 * Math.PI * this.radius;
    }
}

// Subclass for Rectangle
class Rectangle extends Shape {
    constructor(width, height) {
        super("Rectangle");
        this.width = width;
        this.height = height;
    }

    area() {
        return this.width * this.height;
    }

    perimeter() {
        return 2 * (this.width + this.height);
    }
}

// Subclass for RightTriangle
class RightTriangle extends Shape {
    constructor(base, height) {
        super("RightTriangle");
        this.base = base;
        this.height = height;
    }

    area() {
        return 0.5 * this.base * this.height;
    }

    perimeter() {
        // Calculate hypotenuse using Pythagoras theorem
        let hypotenuse = Math.sqrt(this.base * this.base + this.height * this.height);
        return this.base + this.height + hypotenuse;
    }
}

// Creating instances of the shapes
let circle = new Circle(5);  // Radius 5
let rectangle = new Rectangle(4, 6);  // Width 4, Height 6
let triangle = new RightTriangle(3, 4);  // Base 3, Height 4

// Calculating and displaying areas and perimeters
console.log("Circle - Area:", circle.area().toFixed(2), "Perimeter:", circle.perimeter().toFixed(2));
console.log("Rectangle - Area:", rectangle.area(), "Perimeter:", rectangle.perimeter());
console.log("RightTriangle - Area:", triangle.area(), "Perimeter:", triangle.perimeter().toFixed(2));
```

**Problem Statement 7: Implement a Library System**

**Create a class called Library that can manage a collection of books. Each book should be represented as an instance of a Book class (with properties like title, author, and availability). Implement methods in the Library class to add books, check out books, and return books. Keep track of the books available.**

-  


```
// Book class
class Book {
    constructor(title, author) {
        this.title = title;
        this.author = author;
        this.isAvailable = true; // true means the book is available
    }
}

// Library class
class Library {
    constructor() {
        this.books = [];
    }

    // Add a book to the library
    addBook(book) {
        this.books.push(book);
        console.log(`Added "${book.title}" by ${book.author}`);
    }

    // Check out a book by title
    checkOutBook(title) {
        const book = this.books.find(b => b.title === title);
        if (book) {
            if (book.isAvailable) {
                book.isAvailable = false;
                console.log(`You have checked out "${book.title}"`);
            } else {
                console.log(`Sorry, "${book.title}" is not available`);
            }
        } else {
            console.log(`Book titled "${title}" not found`);
        }
    }

    // Return a book by title
    returnBook(title) {
        const book = this.books.find(b => b.title === title);
        if (book) {
            if (!book.isAvailable) {
                book.isAvailable = true;
                console.log(`You have returned "${book.title}"`);
            } else {
                console.log(`"${book.title}" was not checked out`);
            }
        } else {
            console.log(`Book titled "${title}" not found`);
        }
    }

    // Display available books
    displayAvailableBooks() {
        console.log("Available Books:");
        this.books.forEach(b => {
            if (b.isAvailable) {
                console.log(`- "${b.title}" by ${b.author}`);
            }
        });
    }
}

// ----------------------------
// Example Usage

// Create library instance
const myLibrary = new Library();

// Create book instances
const book1 = new Book("1984", "George Orwell");
const book2 = new Book("To Kill a Mockingbird", "Harper Lee");
const book3 = new Book("The Great Gatsby", "F. Scott Fitzgerald");

// Add books to library
myLibrary.addBook(book1);
myLibrary.addBook(book2);
myLibrary.addBook(book3);

// Display available books
myLibrary.displayAvailableBooks();

// Check out a book
myLibrary.checkOutBook("1984");

// Try to check out the same book again
myLibrary.checkOutBook("1984");

// Return the book
myLibrary.returnBook("1984");

// Display available books after return
myLibrary.displayAvailableBooks();
```
